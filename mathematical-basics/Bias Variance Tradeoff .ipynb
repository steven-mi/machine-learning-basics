# Targetfunktion $t(x)$

$$
t: [0,2\pi] \rightarrow R \; with \; t(x) = sin(x)
$$

```python
n = 10000
x = np.linspace(0, 2 * np.pi, n)
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

## Beispiele rausnehmen

```python
X = np.random.uniform(0, 2*np.pi, (2, n))
Y = np.sin(X)
print('X:', X, ' and shape:', X.shape)
print('Y:', Y, ' and shape:', Y.shape)
```

# Calculate hypothesis

## Hypothesis $H_0: h = \theta$

```python
def hypothese_0(theta):
    return lambda x: np.ones_like(x) * theta
```

### All hypothesis
$$
\theta^{D_k} = \frac{\Delta y}{\Delta x} = \frac{y_0 - y_1}{x_0 - x_1} \\
D_k = \left\{ D_1, D_2, D_3,..., D_k\right\} = \\ \left\{ \left\{ (x_0;y_0),(x_k,y_k) \right\} , \left\{ (x_0;y_0),(x_1,y_1) \right\} , \left\{ (x_0;y_0),(x_1,y_1) \right\} , ..., \left\{ (x_0;y_0),(x_1,y_1) \right\} \right\}
$$

```python
def calculate_all_theta(x, y):
    #deltax = x[1] - x[0]
    deltay = y[1] - y[0]
    theta = deltay / 2
    return theta


thetas0 = calculate_all_theta(X, Y)
```

$$
H^{D_k}: h = \theta_k
$$

```python
# activate if you want to plot every hypothesis for every datapoints
for theta in thetas0[:1000]:
    plt.plot(x, hypothese_0(theta)(x))
plt.plot(x, np.sin(x))
plt.show()
```

### Average hypothesis

$$
\tilde{h}(x) = \frac{1}{k} \sum_{i=1}^k H^{D_k} = \frac{1}{k} \sum_{i =1}^k \theta_i
$$

```python
def caculate_average_theta(thetas0):
    theta = np.mean(thetas0)
    return theta


average_theta = caculate_average_theta(thetas0)
plt.plot(x, hypothese_0(average_theta)(x))
plt.plot(x, np.sin(x))
plt.show()
```

## Hypothesis $H_1: h = \theta_0 + \theta_1 x$

```python
def hypothese_1(theta_0, theta_1):
    return lambda x: theta_0 + theta_1 * x
```

### All hypothesis
$$
\theta^{D_k}_1 = \frac{\Delta y}{\Delta x} = \frac{y_0 - y_1}{x_0 - x_1} \\
\theta^{D_k}_0 = y_0 - \theta_1^{D_k} x_0 = y_1 - \theta_1^{D_k} x_1\\
D_k = \left\{ D_1, D_2, D_3,..., D_k\right\} = \\ \left\{ \left\{ (x_0;y_0),(x_k,y_k) \right\} , \left\{ (x_0;y_0),(x_1,y_1) \right\} , \left\{ (x_0;y_0),(x_1,y_1) \right\} , ..., \left\{ (x_0;y_0),(x_1,y_1) \right\} \right\}
$$

```python
def calculate_all_theta(x, y):
    deltax = x[1] - x[0]
    deltay = y[1] - y[0]
    theta_1 = deltay / deltax
    theta_0 = y[1] - theta_1 * x[1]
    return theta_0, theta_1


thetas = calculate_all_theta(X, Y)
```

$$
H^{D_k}: h = \theta^{D_k}_0 + \theta^{D_k}_1 x
$$

```python
# activate if you want to plot every hypothesis for every datapoints
t0 = thetas[0][:1000]
t1 = thetas[1][:1000]

for theta0, theta1 in zip(t0, t1):
    plt.plot(x, hypothese_1(theta0, theta1)(x))
plt.plot(x, np.sin(x))
plt.show()
```

### Average hypothesis

$$
\tilde{h}(x) = \frac{1}{k} \sum_{i=1}^k H^{D_k}(x)
$$

```python
def caculate_average_hypothesis(thetas, x):
    y = []
    t0 = thetas[0]
    t1 = thetas[1]

    for theta0, theta1 in zip(t0, t1):
        y.append(hypothese_1(theta0, theta1)(x))
    y = np.array(y)
    y_average = np.mean(y, axis=0)
    return y_average


y_h1 = caculate_average_hypothesis(thetas, x)
plt.plot(x, y_h1)
plt.plot(x, np.sin(x))
plt.show()
```

    &= variance &+&bias
  \end{align*}

## Bias and variance model 1

```python
bias1 = np.mean((average_theta - np.sin(x)) ** 2)
bias1
```

```python
variance1 = np.mean((thetas0 - average_theta) ** 2)
variance1
```

## Bias and variance model 2

```python
bias2 = np.mean((y_h1 - np.sin(x)) ** 2)
bias2
```

```python
t0 = thetas[0]
t1 = thetas[1]
hd = []
for theta0, theta1 in zip(t0, t1):
    hd.append(hypothese_1(theta0, theta1)(x))    
hd = np.array(hd)

Ed = np.mean((hd - y_h1) ** 2, axis= 1)

variance2 = np.mean(Ed)
variance2
```

# Calculate $E_{out}$
## Model 1

```python
variance1 + bias1
```

## Model 2

```python
variance2 + bias2
```
