# Prerequisites

```python
import numpy as np

import matplotlib.pyplot as plt
```

# Logistic function

## Logistic function - 2d

$$
f(x) = \frac{1}{1+e^{-x}}
$$

```python
def log_func(x): return 1 / (1 + np.exp(-x))


X_log = np.linspace(-5, 5, 1000)
y_log = log_func(X_log)
```

## Plotting logistic function - 2d

```python
plt.plot(X_log, y_log)
plt.title('logistic function - 2d')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

## Logistic function - nd

$$
g_\vec{\theta_{}} (X) = X * \vec{\theta_{}}
\; with \; X\in  \mathbb{R}^{n \times 3} \; and \; \vec{\theta_{}}\in \mathbb{R}^{3 \times 1}$$

$$
\Rightarrow g_i = \sum_{k = 0} x_{ik} \cdot \theta_k
$$

$$ Note: \; X = \begin{bmatrix} \vec{x_{0}} , \vec{x_{1}}, \vec{x_{2}}\end{bmatrix} \; with \; \vec{x_{0}} = \begin{bmatrix} 1 \\ 1 \\\vdots \\ 1 \\ 1\end{bmatrix} \; and \; \vec{x_{0}} \in \mathbb{R}^{n \times 1}  $$

```python
def linear_hypothesis(theta):
    return lambda x: np.dot(x, theta)
```

$$ 
h_{\vec{\theta_{}}} (X) = \frac{1}{1+e^{-g_{\vec{\theta_{}}}(X)}} = \frac{1}{1+e^{-X * \vec{\theta_{}}}}
$$

```python
def logistic_hypothesis(theta):
    return lambda x: 1 / (1 + np.exp(-linear_hypothesis(theta)(x)))
```

```python
X_test = np.random.uniform(-1, -1, (50, 3))
theta_test = np.array([1.1, 2.0, -.9])
h_test = logistic_hypothesis(theta_test)
print(h_test(X_test))
print(h_test(X_test).shape)
```

# Creating dataset

## Trainings data

```python
# class 0:
# covariance matrix and mean
cov0 = np.array([[5, -3], [-3, 3]])
mean0 = np.array([2., 3])

# number of data points
m0 = 1000

# generate m0 gaussian distributed data points with
# mean0 and cov0.
r0 = np.random.multivariate_normal(mean0, cov0, m0)

# covariance matrix
cov1 = np.array([[5, -3], [-3, 3]])
mean1 = np.array([1., 1])
m1 = 1000
r1 = np.random.multivariate_normal(mean1, cov1, m1)
```

```python
plt.scatter(r0[..., 0], r0[..., 1], c='b', marker='o', label="Klasse 0")
plt.scatter(r1[..., 0], r1[..., 1], c='r', marker='x', label="Klasse 1")
plt.xlabel("x0")
plt.ylabel("x1")
plt.show()
```

```python
X = np.append(np.ones((m0 * 2, 1)), np.concatenate((r0, r1)), axis=1)
y = np.zeros(len(r0)+len(r1))
y[:len(r0), ] = 1
```

# Loss functions

## Mean squared loss

$$
L ( \vec { \theta_{} } )_i = \left( h _ { \vec{\theta_{}} } \left( X \right)_i - y_i \right) ^ { 2 }
$$

```python
def mean_squared_loss(x, y, hypothesis):
    return lambda theta: (hypothesis(theta)(x) - y) ** 2


theta = np.array([1.1, 2.0, -.9])
h = logistic_hypothesis
loss_0 = mean_squared_loss(X, y, h)
print(loss_0(theta).shape)
```

## Cross entropy loss


$$
L ( \vec { \theta_{} } )_i =  y_i \;log(h_{\vec{\theta_{}}}(X))_i + (1 - y_i) \; log(1 - h_{\vec{\theta_{}}}(X))_i
$$

```python
def cross_entropy_loss(x, y, hypothesis):
    return lambda theta: y * np.log(hypothesis(theta)(x)) + (1 - y) * np.log(1 - hypothesis(theta)(x))
```

```python
loss_1 = cross_entropy_loss(X, y, h)
print(loss_1(theta).shape)
```

# Cost function

$$
C(\vec{\theta_{}}) = -
\frac { 1 } { m } \sum _ { i = 1 } ^ { m } L(\vec{\theta_{}})_i
$$

```python
def cost_function(X, y, h, loss):
    return lambda theta: -1 / len(theta) * np.sum(loss(X, y, h)(theta),  axis=0)
```

```python
print('cost function with cross entropy loss',
      cost_function(X, y, h, cross_entropy_loss)(theta))
print('cost function with mean squared loss',
      cost_function(X, y, h, mean_squared_loss)(theta))
```

# Gradient descent

## Sigmoid derivative

$$
f(x) = \frac{1}{1+e^{-x}}
$$

$$
f^\prime(x) = \frac{e^{-x}}{(1+e^{-x})^2} = \frac{e^{-x }+1 -1}{1+e^{-x}} \frac{1}{1+e^{-x}}= \\ 
\left(1 - \frac{1}{1+e^{-x}}\right) \frac{1}{1+e^{-x}} = (1 - f(x)) f(x)
$$

## Gradient
$$
\frac{\partial C(\vec{\theta_{}})}{\partial \theta_j}=
\frac{\partial \frac { -1 } { m } \sum _ { i = 1 } ^ { m } y_i \;log(h_{\vec{\theta_{}}}(X))_i + (1 - y_i) \; log(1 - h_{\vec{\theta_{}}}(X))_i}{\partial \theta_j} = \\
\frac{-1}{m} \sum _ { i = 1 } ^ { m } \frac{\partial  y_i \;log(h_{\vec{\theta_{}}}(X))_i}{\partial \theta_j} + \frac{\partial (1 - y_i) \; log(1 - h_{\vec{\theta_{}}}(X))_i}{\partial \theta_j} = \\
\frac{-1}{m} \sum _ { i = 1 } ^ { m } y_i \frac{\partial  \;log(h_{\vec{\theta_{}}}(X))_i}{\partial \theta_j} + (1 - y_i) \frac{\partial log(1 - h_{\vec{\theta_{}}}(X))_i}{\partial \theta_j} = \\
\frac{-1}{m} \sum _ { i = 1 } ^ { m } y_i \frac{1}{h_{\vec{\theta_{}}}(X)_i}\frac{\partial h_{\vec{\theta_{}}}(X)_i}{\partial \theta_j} - (1 - y_i) \frac{1}{1 - h_{\vec{\theta_{}}}(X))_i} \frac{\partial h_{\vec{\theta_{}}}(X)_i}{\partial \theta_j} = \\
\frac{-1}{m} \sum _ { i = 1 } ^ { m } \frac{y_i}{h_{\vec{\theta_{}}}(X)_i}\frac{\partial h_{\vec{\theta_{}}}(X)_i}{\partial \theta_j} - \frac{1 - y_i}{1 - h_{\vec{\theta_{}}}(X))_i} \frac{\partial h_{\vec{\theta_{}}}(X)_i}{\partial \theta_j} = \\
\frac{-1}{m} \sum _ { i = 1 } ^ { m } \frac{y_i}{h_{\vec{\theta_{}}}(X)_i} \frac{1}{h_{\vec{\theta_{}}}(X)_i (1 - h_{\vec{\theta_{}}}(X)_i)}
\frac{\partial g_{\vec{\theta_{}}}(X)_i}{\partial \theta_j} - \frac{1 - y_i}{1 - h_{\vec{\theta_{}}}(X))_i} \frac{1}{ h_{\vec{\theta_{}}}(X)_i (1 - h_{\vec{\theta_{}}}(X)_i)} \frac{\partial g_{\vec{\theta_{}}}(X)_i}{\partial \theta_j}= \\
\frac{-1}{m} \sum _ { i = 1 } ^ { m } \frac{y_i}{h_{\vec{\theta_{}}}(X)_i^2 (1 - h_{\vec{\theta_{}}}(X)_i )}
\frac{\partial \sum_{k = 0} x_{ik} \cdot \theta_k}{\partial \theta_j} - \frac{1 - y_i}{h_{\vec{\theta_{}}}(X)_i (1 - h_{\vec{\theta_{}}}(X)_i )^2} \frac{\partial \sum_{k = 0} x_{ik} \cdot \theta_k}{\partial \theta_j}= \\
\frac{-1}{m} \sum _ { i = 1 } ^ { m } \frac{y_i}{h_{\vec{\theta_{}}}(X)_i^2 (1 - h_{\vec{\theta_{}}}(X)_i )}x_{ij} - \frac{1 - y_i}{h_{\vec{\theta_{}}}(X)_i (1 - h_{\vec{\theta_{}}}(X)_i )^2} x_{ij} = \\
\frac{-1}{m} \sum _ { i = 1 } ^ { m } x_{ij} \left( \frac{y_i}{h_{\vec{\theta_{}}}(X)_i^2(1 - h_{\vec{\theta_{}}}(X)_i )} - \frac{1 - y_i}{h_{\vec{\theta_{}}}(X)_i (1 - h_{\vec{\theta_{}}}(X)_i )^2} \right)= \\
\frac{-1}{m} \sum _ { i = 1 } ^ { m } x_{ij} \left( \frac{y_i - y_i h_{\vec{\theta_{}}}(X)_i }{h_{\vec{\theta_{}}}(X)_i^2(1 - h_{\vec{\theta_{}}}(X)_i )^2} - \frac{h_{\vec{\theta_{}}}(X)_i  - y_ih_{\vec{\theta_{}}}(X)_i }{h_{\vec{\theta_{}}}(X)_i ^2 (1 - h_{\vec{\theta_{}}}(X)_i )^2} \right)= \\
\frac{-1}{m} \sum _ { i = 1 } ^ { m } x_{ij} \left( \frac{y_i - h_{\vec{\theta_{}}}(X)_i }{h_{\vec{\theta_{}}}(X)_i ^2 (1 - h_{\vec{\theta_{}}}(X)_i )^2} \right) =\\ \frac{1}{m} X^T \left( \frac{h_{\vec{\theta_{}}}(X) - \vec{y_{}}}{h_{\vec{\theta_{}}}(X)^2 \odot (1 - h_{\vec{\theta_{}}}(X))^2} \right) \approx \frac{1}{m}X^T(h_{\vec{\theta_{}}}(X) - \vec{y_{}})
$$

## Calculate theta with gradient

```python
def compute_new_theta(x, y, theta, alpha, hypothesis):
    return theta - alpha / len(x) * np.dot(x.T, -y + hypothesis(theta)(x))
```

## Minimizing cost function

```python
def gradient_descent(alpha, theta, nb_iterations, X, y):
    theta_history = []
    for i in range(nb_iterations):
        theta = compute_new_theta(X, y, theta, alpha, logistic_hypothesis)
        theta_history.append(theta)
    return theta_history


thetas = gradient_descent(0.1, theta, 100, X, y)

print('loss at starting point', cost_function(
    X, y, logistic_hypothesis, cross_entropy_loss)(theta))
print('loss after training', cost_function(
    X, y, logistic_hypothesis, cross_entropy_loss)(thetas[-1]))
```

# Entscheidungsebene



```python
opt_theta = thetas[-1]
X1 = X.T[1].T
X2 = X.T[2].T
```

## a

$$
\frac{1}{1+e^{-(\theta_0 + \theta_1 x_1 + \theta_2 x2)}} = 0.5 \\
1 = 0.5 + 0.5e^{-(\theta_0 + \theta_1 x_1 + \theta_2 x2)} \\
1 = e^{-(\theta_0 + \theta_1 x_1 + \theta_2 x2)} \\
0 = -(\theta_0 + \theta_1 x_1 + \theta_2 x2)
\\\vdots \\
equals \; to \; b.
$$

## b

$$
\theta_0 + \theta_1 x_1 + \theta_2 x_2 = 0
\\\Rightarrow x_2 = \frac{-\theta_0 - \theta_1 x_1}{\theta_2}
\\\Rightarrow x_1 = \frac{-\theta_0 - \theta_2 x_2}{\theta_1}
$$

```python
y_ebene_1 = (- X1 * opt_theta[1] - opt_theta[0])/opt_theta[2]
y_ebene_2 = (- X2 * opt_theta[2] - opt_theta[0])/opt_theta[1]
```

```python
plt.figure(figsize=(10, 10))
plt.plot(X1, y_ebene_1, label="Ebene 1", c='y')
plt.plot(y_ebene_2, X2, label="Ebene 2", c='g')
plt.scatter(r0[..., 0], r0[..., 1], c='b', marker='o', label="Klasse 0")
plt.scatter(r1[..., 0], r1[..., 1], c='r', marker='x', label="Klasse 1")
plt.xlabel("x0")
plt.ylabel("x1")
plt.legend()
plt.show()
```

# Classifications error

$$
error_{classification} = \frac{\# \Omega _{prediction}^{(wrong)}}{\#X}
$$

$$
\theta_0 + \theta_1 x_1 + \theta_2 x_2 = 0
\\ \Rightarrow x_2 = \frac{-\theta_0 - \theta_1 x_1}{\theta_2} \\
\Rightarrow x_1 = \frac{-\theta_0 - \theta_2 x_2}{\theta_1}
$$

```python
pred = logistic_hypothesis(opt_theta)(X)

pred_one_hot = pred
pred_one_hot[pred_one_hot >= 0.5] = 1
pred_one_hot[pred_one_hot < 0.5] = 0
pred_wrong = pred_one_hot[y != pred_one_hot]

amount_wrong_pred = len(pred_wrong)
amount_pred = len(pred)

error_classification = amount_wrong_pred / amount_pred
print('classifications error is', error_classification)
```

# Lambda function

## Cost function with lambda regularization

$$
C(\vec{\theta_{}}) =
\frac { 1 } { m } \sum _ { i = 1 } ^ { m } L(\vec{\theta_{}})_i + \frac{1}{2}\sum_{i=1} \lambda \theta_i ^2
$$

```python
def regularization_function(reg_rate):
    return lambda theta: reg_rate * (theta ** 2)
```

```python
def cost_function_reg(X, y, h, loss, regularization, regularization_rate):
    return lambda theta: -1 / len(theta) * (np.sum(loss(X, y, h)(theta),  axis=0) + np.sum(regularization_function(regularization_rate)(theta))/2)
```

## Gradient descent with lambda regularization

$$
\frac{\partial C(\vec{\theta_{}})}{\partial \theta_j}=
\frac{\partial \frac { -1 } { m } \sum _ { i = 1 } ^ { m } L(\vec{\theta_{}})_i + \frac{1}{2} \sum_{i=1} \lambda \theta_i ^2}{\partial \theta_j} = \\
\frac{\partial  \frac { -1 } { m } \sum _ { i = 1 } ^ { m } L(\vec{\theta_{}})_i }{\partial \theta_j} + \frac{\partial  \frac{1}{2} \sum_{i=1} \lambda \theta_i ^2}{\partial \theta_j} = \\
\frac{\partial  \frac { -1 } { m } \sum _ { i = 1 } ^ { m } L(\vec{\theta_{}})_i }{\partial \theta_j} + \lambda  \sum_{i=1}\frac{\partial  \theta_i ^2}{\partial \theta_j}= \\
\frac{\partial  \frac { -1 } { m } \sum _ { i = 1 } ^ { m } L(\vec{\theta_{}})_i }{\partial \theta_j} + 2\lambda \theta_j \approx \frac{1}{m}X^T(h_{\vec{\theta_{}}}(X) - \vec{y_{}}) + \lambda \odot \theta
$$

```python
def compute_new_theta_reg(x, y, theta, alpha, hypothesis, reg_rate):
    return theta - (alpha / len(x) * np.dot(x.T, -y + hypothesis(theta)(x)) + reg_rate * theta)
```

```python
def gradient_descent_reg(alpha, theta, nb_iterations, X, y):
    theta_history = []
    for i in range(nb_iterations):
        theta = compute_new_theta_reg(
            X, y, theta, alpha, logistic_hypothesis, 0.1)
        theta_history.append(theta)
    return theta_history


thetas = gradient_descent_reg(0.1, theta, 100, X, y)

print('loss at starting point', cost_function_reg(
    X, y, logistic_hypothesis, cross_entropy_loss, regularization_function, 0.1)(theta))
print('loss after training', cost_function_reg(
    X, y, logistic_hypothesis, cross_entropy_loss, regularization_function, 0.1)(thetas[-1]))
```
